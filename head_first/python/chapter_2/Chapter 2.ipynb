```python
# odds.py
from datetime import datetime
import time
import random
print(dir(random))

odds = []
for i in range(1, 60):
    if i % 2 != 0:
        odds.append(i)

for i in range(1, 5):
    right_this_minute = datetime.today().minute
    if right_this_minute in odds:
        print("This minute is an odd minute!.")
    else:
        print("This is an even minute :) ")
        wait_time = random.randint(1, 60) # integer
    time.sleep(wait_time)
```

```python
# vowels.py list example
vowels = ['a', 'e', 'i', 'o'] # declaring a list
vowels.append('u') # missed an entry - append keyword
word = 'Milliways'
found_vowels = []
found_consonants = []
print("Vowels:")
for letter in word: # in keyword to iterate over every character in a string
    if letter.lower() in vowels: # in keyword to check an item exists
        found_vowels.append(letter)
print(found_vowels)
print("Consonants:")
for letter in word: # in keyword to iterate over every character in a string
    if letter.lower() not in vowels: # in keyword to check an item exists
        found_consonants.append(letter)
print(found_consonants)
```

```python
# other list keywords
l = list()
l.append(0)
l.append(1)
l.append(2)
l.append(3)
l.append(4)
l.append(5)
l.append(5)
print(l)
l.remove(5) # removes specified item exactly from the list
print(l)
l.pop() # pops the top item from the list - index not specified - STACK
print(l)
l.pop(0) # pops the item at index 0 from the list - index specified
print(l)
l.extend([6, 7, 8, 9]) # joins two lists together, joined at last position
print(l)
l.insert(4, 5) # add at index(n-1) 4 the number 5
print(l)
```

```python
# list copy
# examples when the copied object refers to the same object
first = [1, 2, 3, 4, 5]
second = first
print(first, second, id(first), id(second))
second.append(6)
print(first, second, id(first), id(second))
# example when the copied object creates a new object and stores it
third = second.copy()
print(third, id(first), id(second), id(third))
```

```python
# negative indices - start from -1 to -n
print(first[-1])
# access each item in the list directly as follows
print(first[3], first[-3])
# step wise in list
print(first[0:len(first)-1:2])
# slices
print(first[:3])
print(first[3:])
```
